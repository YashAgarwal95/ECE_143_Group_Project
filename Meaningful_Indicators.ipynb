### Basic properties
We identified a dataset "Education Statistics" from The World Bank to study and describe its basic properties. The data including geographical areas, income of each areas, country names, indicator names and data range from 1970 to 2014.

We are going to explore the data to catch some meaningful indicators within 217 countries.

```python
import pandas as pd
df = pd.read_csv("EdStatsData.csv")
```

Columns of Indicator Name, Indicator Code and data range from 1970 to 2014 are selected. While indicators without data are ignored, those indicators with more data are sorted in descending order. A sub-dataframe is created for above purpose.

```python
df2 = df.iloc[:,4:49].notnull().astype(int)
df2 = pd.concat([df['Indicator Name'], df['Indicator Code'],df2],axis=1)
df2 = df2.iloc[91625:886930,:]
df2['Total'] = df2.sum(axis=1)
```

```python
df2
```

```python
df2.groupby('Indicator Name')['Total'].sum()
grouped = df2.groupby('Indicator Name')['Total'].sum().reset_index().sort_values('Total', ascending=False)
```

On the data below, it shows high occurence of enrolment ratio, literacy rate, educations completion rates, unemployment, GDP, GNI, expenditure, etc. This could be used to analyze how much wealth contributes to education.

```python
grouped
```

```python
grouped.to_csv('grouped.csv',sep=',')
```
